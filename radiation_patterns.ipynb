The geometry of the radiation power or fields of a radiating system can be visualized by plotting a "radiation pattern". This usually means showing how some normalized version of the electric field or Poynting vector in 2d or 3d looks like as function of angles. Below we will define the radiation pattern, $|F(\theta,\phi)|^2$ by 
$$
|F(\theta,\phi)|^2 = \frac{r^2 (\hat{r}\cdot\vec{S})}{P_\text{tot}} 
$$
So roughly, it is just a rescaling of the Poynting vector. The radiation pattern is plotted in parametric plots as a function of $\theta$ and $\phi$, where the distance of the surface or curve away from the origin is simply $|F(\theta,\phi)|^2$.

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as axes3d
```

# Dipole Radiation Pattern 
Here we have 
$$\vec{S}=\frac{\mu_0}{16 \pi^2 c}|\ddot{p}|^2 \frac{\sin ^2 \theta}{r^2} \hat{r}, $$ 
so the radiation pattern is
$$|F(\theta,\phi)|^2 \propto \sin^2\theta$$

```python
%matplotlib ipympl
theta, phi = np.linspace(0, np.pi, 200), np.linspace(0, 2 * np.pi, 200)
THETA, PHI = np.meshgrid(theta, phi)
Ptot = 1./(6.*np.pi)
R = np.power(np.sin(THETA),2)/(16.*np.pi**2) / Ptot
X = R * np.sin(THETA) * np.cos(PHI)
Y = R * np.sin(THETA) * np.sin(PHI)
Z = R * np.cos(THETA)
fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection='3d')
plot = ax.plot_surface(
    X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('jet'),
    linewidth=0, antialiased=False, alpha=0.5)
```

# Radiation pattern of an accelerating charge with $\vec{a} \parallel \vec{v}$
Here we the radiated power contains an additional factor of $(1-\vec{v}\cdot \hat{r}/c)$ as we discussed in class, so 

$$|F(\theta,\phi)|^2 \propto \frac{\sin ^2 \theta}{(1-(v/c)\cos \theta)^5}$$

```python
%matplotlib ipympl
theta, phi = np.linspace(0, np.pi, 200), np.linspace(0, 2 * np.pi, 200)
THETA, PHI = np.meshgrid(theta, phi)
#R = np.cos(THETA**2)
beta = 0.9 #v/c
Ptot = 8.*np.pi/3. * np.power(1/np.sqrt(1-beta**2),6)
R = np.power(np.sin(THETA),2)/np.power(1-beta*np.cos(THETA),5) / Ptot
X = R * np.sin(THETA) * np.cos(PHI)
Y = R * np.sin(THETA) * np.sin(PHI)
Z = R * np.cos(THETA)
fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection='3d')
plot = ax.plot_surface(
    X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('jet'),
    linewidth=0, antialiased=False, alpha=0.5)
plt.title(r"$\beta = "+str(beta)+r"$")
```

# Radiation pattern of an accelerating charge with $\vec{a} \perp \vec{v}$, (synchrotron for $\vec{v}\propto \hat{z}$, $\vec{a}\propto\vec{x}$
We have not derived this angular distributio in class, but you can find details in graduate textbooks, e.g. Zangwill 23.3.3
$$|F(\theta,\phi)|^2 \propto \frac{1}{(1-\beta \cos \theta)^3}\left[1-\frac{\sin ^2 \theta \cos ^2 \phi}{\gamma^2(1-\beta \cos \theta)^2}\right]$$

```python
%matplotlib ipympl
theta, phi = np.linspace(0, np.pi, 200), np.linspace(0, 2 * np.pi, 200)
THETA, PHI = np.meshgrid(theta, phi)
#R = np.cos(THETA**2)
beta = 0.9#v/c
g = 1/np.sqrt(1-beta**2)
Ptot = 8.*np.pi/3. * np.power(g,4)
R = (1. - np.power(np.sin(THETA),2)*np.power(np.cos(PHI),2)/(g**2 * np.power(1. - beta*np.cos(THETA),2)))/np.power(1-beta*np.cos(THETA),3) / Ptot
X = R * np.sin(THETA) * np.cos(PHI)
Y = R * np.sin(THETA) * np.sin(PHI)
Z = R * np.cos(THETA)
fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection='3d')
plot = ax.plot_surface(
    X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('jet'),
    linewidth=0, antialiased=False, alpha=0.5)
plt.title(r"$\beta = "+str(beta)+r"$")
plt.xlim([-5,5])
plt.ylim([-5,5])
```

```python
theta = np.linspace(0, np.pi, 200)

#R = np.cos(THETA**2)
beta = 0.95 #v/c


#fig = plt.figure()
#ax = fig.add_subplot(1,1,1, projection='3d')
plt.figure()
styles = ['-','--', '-.',':']
for i, beta in enumerate([0, 0.5, 0.9,0.95]):

    g = 1/np.sqrt(1-beta**2)
    Ptot = 8.*np.pi/3. * np.power(g,4)
    PHI = 0.*np.pi/2
    R = (1. - np.power(np.sin(theta),2)*np.power(np.cos(PHI),2)/(g**2 * np.power(1. - beta*np.cos(theta),2)))/np.power(1-beta*np.cos(theta),3) / Ptot
    X = R * np.sin(theta)
    Z = R * np.cos(theta)
    p = plt.plot(X,Z, label=r'$\beta = $'+str(beta), ls=styles[i],lw=2)
    plt.plot(-X,Z,ls=styles[i],lw=2,c=p[0].get_color())
    
plt.xlabel('x')
plt.ylabel('z')
plt.xlim([-5,5])
plt.ylim([0,10])
plt.legend()
plt.title(r'Radiation Pattern of a Charge with $\mathbf{v} \perp \mathbf{a}$,  $\mathbf{v}\propto\hat{\mathbf{z}} $ ')
```
